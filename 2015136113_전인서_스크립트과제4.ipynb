#### - 1) 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s1.txt에 있는 파일 내용은 다음과 같다.

>cat dog  
 dog pig  
 ham bird  
 pig ham  

```python
with open('s.txt', 'r') as f:
    s = f.read()
    a=s.split('\n')
    a.sort()
    s1 = '\n'.join(a)
with open('s1.txt','w') as f:
    f.write(s1)
with open('s1.txt','r') as f:
    s1 = f.read()
    print(s1)
```

---
우선 s.txt에 있는 텍스트를 읽어서 s에 저장하고 s를 \n 를 기준으로 split 해주고 정렬을 해준다. 그리고 그 값을 '\n'.join 을 써서 다시 합쳐서 문자열s1에 넣어준다.  
그후 s1.txt를 쓰기형식으로 열어서 문자열s1을 저장해준다.

#### - 2) 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s2.txt에 있는 파일 내용은 다음과 같다.
>ham bird  
 cat dog  
 pig ham  
 dog pig  

```python
def cmp_by_each_element(a):
    return a[1]

with open('s.txt', 'r') as f:
    s = f.read()
    a=s.split('\n')
    l = []
    l2 = []
    for k in a:
        l.append(k.split(' '))
    l.sort(key = cmp_by_each_element)
    for k in l:
        l2.append(' '.join(k))
    s2 = '\n'.join(l2)
with open('s2.txt','w') as f:
    f.write(s2)
with open('s2.txt','r') as f:
    s2 = f.read()
    print(s2)
```

---
우선 리스트의 1번째 인덱스를 반환하는 함수인 cmp_by_each_element()를 정의해준다.  
s.txt파일을 읽기형식으로 열어서 문자열 s에 텍스트를 넣어준다. 1-1번과 같이 \n를 기준으로 split해 a에 넣는다.  
for문을 써서 빈 리스트 l에 a의 인덱스 하나하나마다 다시 공백을 기준으로 split한 값을 넣어준다.
그리고 리스트 l을 cmp_by_each_element()이용해 두번째 단어를 기준으로 정렬해준다.  
그후 다시 for문을 이용해 l2리스트에 l의 인덱스를 공백을 기준으로 join한 값을 넣어준다.  
문자열 s2에 l2를 '\n'join() 한값을 넣어준다.  
s2.txt를 쓰기로 열어 s2값을 넣어준다.

#### - 3) 파일 s.txt을 읽어서 각 라인들에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라인에 세 개의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s3.txt에 있는 파일 내용은 다음과 같다.
 >pig ham cat  
 dog ham bird  
 dog pig  

```python
with open('s.txt', 'r') as f:
    s = f.read()
    a = s.split()
    l=[]
    for x in range(len(a)):
        if((x+1)%3==0):
            l.append(a[x-2]+' '+a[x-1]+' '+a[x])
    if(len(a)%3 ==1):
        l.append(a[-1])
    if(len(a)%3 ==2):
        l.append(a[-2]+' '+a[-1])
    s3 = '\n'.join(l)
with open('s3.txt','w') as f:
    f.write(s3)
with open('s3.txt','r') as f:
    s3 = f.read()
    print(s3)
```

---
s.txt를 읽어와 s에 텍스트를 넣어준다.   
a에 s값을 단어마다 split해서 넣어준다.  
for문을 사용해서 리스트 a의 크기만큼 반복되게 해서 만약 a의 크기만큼 점점 x가 커질 때 x+1을 3으로 나눈 나머지가 0, 즉 3의 배수가 되면 빈 리스트 l에 x의 값과 그 전값과 그 전전값을 같이 넣어준다.  
이렇게 하면 맨 마지막 단어가 1개나 2개가 남아있으면 출력을 못해주므로 a의 길이를 3으로 나눴을 때 1,2가 되면 각각 마지막 1개, 2개의 단어를 l에 넣어준다.  
s3에 l을 \n으로 join해준 값을 넣어준다.  
그 후 s3.txt를 쓰기형식으로 열어 s3의 문자열을 넣어준다.

#### [3번 문제]

range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.  
•1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.  
>range(stop)
 range(start, stop[, step])

•2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.  
•3) frange 사용 예  
◾3-1) frange(0.5)◦[0.0, 0.1, 0.2, 0.3, 0.4]  

◾3-2) frange(1.0, 2.0)◦[1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]  

◾3-3) frange(2.2, 4.0, 0.5)◦[2.2, 2.7, 3.2, 3.7]  

```python
def frange(stop ,start=0.0, step=0.1):
    l = []
    if(start>stop):
        start,stop = stop,start
    for x in range(int(start*10),int(stop*10),int(step*10)):
        l.append(x/10)
    return l
print(frange(0.5))
print(frange(1.0,2.0))
print(frange(2.2,4.0,0.5))
```

---
우선 frange의 매개변수로 stop은 입력이 필수적이므로 냅두고, start와 step의 디폴트값을 0.0, 0.1로 정해준다.  
만약 start값이 stop값보다 크다면 사용자가 start,stop순서로 값을 입력한 것이므로 두 값을 바꾸어 준다.  
range는 정수만 들어가 지므로 각각의 값을 10으로 곱해서 int형으로 바꾸어준후 그 증가되는 값들을 10으로 다시 나누어 빈리스트 l에 넣어준다.  
그리고 l을 반환해준다.

◾sum(1, 2, 3, 4, 5)  
    ◦15  

◾sum(1, 5, 7, 2, -10)

```python
def sum(a=0,*var):
    for num in var:
        a = a + num
    return a
print(sum())
print(sum(1,2))
print(sum(1,2,3,4,5))
print(sum(1,5,7,2,-10))
```

#### [5번 문제]

여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.  
•1) 다음에 제시되는 함수들을 모두 이용해야 한다.  
◾split  
◾map  
◾join  

•2) myinitial() 함수 사용  예  
◾myinitial("as soon as possible")  
◦ASAP  




```python
def myinitial(s):
    l=s.split()
    l2 =[]
    for a in l:
        l2.append(a[0])
    Y = list(map(lambda x: x.capitalize(), l2))
    s2=''.join(Y)
    return s2

print(myinitial("as soon as possible"))
```

---
myinitial에 문자열 s를 입력받으면 그값을 단어로 split해서 문자열 l에 넣어준다. 그 각각의 단어들의 첫번째 문자를 다시 리스트 l2에 넣어준다. 
map을 써서 lambda를 쓴 x의 문자를 대문자로 바꾸어주는 것을 l2에 적용한 값을 리스트로 바꾸어주어 그 값을 리스트 Y에 넣어준다.   
그 후 s2에 ''.join(Y)을 쓴 값을 넣어주고 s2를 반환해준다.

#### [6번 문제]

음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.  
•[옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.   


```python
def myfact(n):
    if n == 1:
        return 1
    else:
        return(n+myfact(n-1))
    
print(myfact(1000))
```

---
음이 아닌 정수 n을 입력받으면 종료조건으로 만약 n이 1이라면 1을 반환하게한다.  
아니라면 재귀함수를 이용해 n+myfact(n-1)을 반환하게 한다.  
n이 1이 될 때까지 반복되게 된다.

우선 import string은 가장 기본적인 형태이다. 이름 string이 그대로 유지되므로 string.split() 형태로 자격이름을 사용하게된다. 그리고 보통의 문이 생성될 수 있는 곳이라면 어디서든 사용될 수 있다. def문이나 if문에서도 사용될 수 있다.  

이 와는 달리 from string import* 는 해당 모듈에 존재하는 '__ ' 로 시작되는 이름들을 제외한 모든 이름들을 현재 이름 공간으로 불러들인다.
불러들인 각 이름들은 모듈 이름 없이 직접 사용 가능하다. 그래서 string.split()으로 쓰이지 않고 split의 형태로 사용하게 된다. 또 import 하는 이름들이 기존에 미리 존재하고 있었다면 그 이름들에 의해 참조되던 기존 객체들은 상실된다.

•7) [주의] 필수사항  
◾위에서 만든 검색엔진 코딩은 매우 간단한 것이라 부족한 점이 많이 존재한다.  
◾본인이 생각하기에 상업적인 완성도 높은 검색 로봇/엔진이 되려면 어떤 기능들이 추가적으로 구현되어야 할지 최소 1가지 이상 제시하시오.  



```python
import requests
import re
import string  

def call(b):  
    req = requests.get(b)
    source = req.text
    
    text1 = re.sub('<script.*?>.*?</script>', ' ', source, 0, re.I|re.S)
    text2 = re.sub('<style.*?>.*?</style>', ' ', text1, 0, re.I|re.S)
    text3 = re.sub('<.+?>', ' ', text2, 0, re.I|re.S)
    text4 = re.sub('-->', ' ', text3, 0, re.I|re.S)
    text5 = re.sub("[!|?|<|>|:|\[|\]|#|$|%|&|\(|\)|*|+|\-|,|.|/|;|=|@|^|_|`|{|}|~|\"|\'|\\\|/]",' ', text4, 0, re.I|re.S)
    s=text5.split()
    dic={}
    j=0
    for i in s:
        if i in dic:
            dic[i] += 1
        else:
            dic[i] = 1
    print(dic)

print('http://cse.koreatech.ac.kr')    
call('http://cse.koreatech.ac.kr')
print()

print('https://www.koreatech.ac.kr')
call('https://www.koreatech.ac.kr')
print()

print('http://www.naver.com')
call('http://www.naver.com')
print()

print('http://www.daum.net')
call('http://www.daum.net')
print()

print('http://www.nytimes.com')
call('http://www.nytimes.com')
```
