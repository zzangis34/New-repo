### •5. 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.  
◾1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)◦>>> addall([1])  
1  

◦>>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])  
45  


◾2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)◦>>> addallodd([1])  
1  

◦>>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])  
25  


◾[참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)  
◾[참고 2]: sum() 등의 내장함수 사용 가능함  


```python
def addall(L):
    sum = 0
    for k in L:
        sum += k
    return sum
print(addall([1]))
print(addall([1,2,3,4,5,6,7,8,9]))
```

```python
def addall(L):
    return sum([k for k in L])
print(addall([1]))
print(addall([1,2,3,4,5,6,7,8,9]))
```

---
우선 위에 함수는 for in을 써서 구현하였는데, sum을 0으로 초기화하고 for in을 이용해서 sum에 리스트 원소 값을 모두 더해준다.  
그리고 그 값을 반환한다.  

아래 함수는 리스트 내포 방식을 써서 여러줄로 썼던 것을 한 줄로 표현하였는데, 내장함수 sum을 활용해서 sum안에 리스트L의 원소 들을 더해주고 반환한다.

---

```python
def addallodd(L):
    sum = 0
    for k in L:
        if k%2 == 1:
            sum += k
    return sum
print(addallodd([1]))
print(addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

```python
def addallodd(L):
    return sum([k for k in L if k%2 ==1])
print(addallodd([1]))
print(addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

---
위에 함수는 sum을 0으로 초기화한 후, for in을 사용해서 모든 수 중에서 홀수를 찾기위해 k%2가 1이 되는 수를 찾아 sum에 모두 더한 후 반환 한다.  
밑에 함수는 내장함수 sum을 이용해서 리스트 내포방식으로 한줄로 정리해서 알고리즘은 위 함수와 똑같이 반환한다.

### •[ACM-ICPC 문제 1673번]  
◾치킨 쿠폰  


```python
import sys

while 1:
    line = sys.stdin.readline().rstrip('\n')
    if not line:
        break
    else:
        n, k = map(int, line.split(" "))
        max = n
        x=n
        while k<=x:
            max += x//k
            x = x//k + x%k
        print(max)
```

---
x가 쿠폰의 도장 수도장수, max는 최대 치킨수이고 n장의 쿠폰이 있으면 n마리 만큼의 치킨을 먹을 수 있다.  
n마리의 치킨을 먹으면 n개의 도장이 찍힌다.  
그래서  x = n 이 된다.  도장 수 x가 k 보다 적으면 먹을수 없을때까지 반복한다.  
치킨을 먹은수(a//k)와 사용하지 않은 도장수(x%k)의 합은 max 에 도장수에서 k로 나눈값을 더하고 도장수이다.

◦예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}   


  
◾요구사항 4. 다음 5개의 사이트에 대한 단어 출현빈도 결과를 출력하시오.  
 

```python
import requests
import re
import string  

def call(b):  
    reqest = requests.get(b)
    source = reqest.text
    text1 = re.sub('<script.*?>.*?</script>', ' ', source, 0, re.I|re.S)
    text2 = re.sub('<style.*?>.*?</style>', ' ', text1, 0, re.I|re.S)
    text3 = re.sub('<.+?>', ' ', text2, 0, re.I|re.S)
    text4 = re.sub('-->', ' ', text3, 0, re.I|re.S)
    text5 = re.sub("[!|?|<|>|:|\[|\]|#|$|%|&|\(|\)|*|+|\-|,|.|/|;|=|@|^|_|`|{|}|~|\"|\'|\\\|/]",' ', text4, 0, re.I|re.S)
    s=text5.split()
    dic={}
    j=0
    for i in s:
        if i in dic:
            dic[i] += 1
        else:
            dic[i] = 1
    print(dic)

print('http://cse.koreatech.ac.kr')    
call('http://cse.koreatech.ac.kr')
print()

print('https://www.koreatech.ac.kr')
call('https://www.koreatech.ac.kr')
print()

print('http://www.naver.com')
call('http://www.naver.com')
print()

print('http://www.daum.net')
call('http://www.daum.net')
print()

print('http://www.nytimes.com')
call('http://www.nytimes.com')
```

---
각 웹페이지의 html을 텍스트로 받아와 각각의 text에 저장한다.  
script, style, html태그, 주석, punctuation을 공백으로 바꾸고, 바뀌어진 문자열을 공백단위로 쪼개어 리스트에 넣는다.   
반복문을 이용해서 사전에 같은 단어가 없으면 value를 1로 하고, value를 1씩 증가시켜준다.
